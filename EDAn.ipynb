Exploratory Data Analysis (EDA) is a method of analyzing datasets to understand their main characteristics. It involves summarizing data features, detecting patterns, and uncovering relationships through visual and statistical techniques. EDA helps in gaining insights and formulating hypotheses for further analysis.

```python
#1. How do vaccination rates correlate with a decrease in disease incidence?

# Merge the vaccination coverage and disease incidence data based on year and region
merged_data = pd.merge(data1, data2, on=['YEAR', 'NAME'], how='inner')


# Calculate the correlation between coverage and incidence rates.
correlation = merged_data['COVERAGE'].corr(merged_data['INCIDENCE_RATE'])

print(f"Correlation between vaccination coverage and disease incidence rate: {correlation}")

# Visualize the relationship.
plt.figure(figsize=(8, 6))
sns.scatterplot(x='COVERAGE', y='INCIDENCE_RATE', data=merged_data)
plt.xlabel('Vaccination Coverage (%)')
plt.ylabel('Disease Incidence Rate')
plt.title('Correlation between Vaccination Coverage and Disease Incidence')
plt.show()

#disease-specific correlation (replace 'Measles' with other diseases):
for disease in merged_data['DISEASE'].unique():
    disease_data = merged_data[merged_data['DISEASE'] == disease]
    disease_correlation = disease_data['COVERAGE'].corr(disease_data['INCIDENCE_RATE'])
    print(f"Correlation for {disease}: {disease_correlation}")
```

```python
#2. What is the drop-off rate between 1st dose and subsequent doses?

# Define dose pairs from column "ANTIGEN_DESCRIPTION"
dose_pairs = [
    ('DTP-containing vaccine, 1st dose', 'DTP-containing vaccine, 3rd dose'),
    ('Inactivated polio-containing vaccine, 1st dose', 'Inactivated polio-containing vaccine, 2nd dose'),
    ('Fractional inactivated polio-containing vaccine, 1st dose', 'Fractional inactivated polio-containing vaccine, 2nd dose'),
    ('Measles-containing vaccine, 1st dose', 'Measles-containing vaccine, 2nd dose'),
    ('Japanese encephalitis vaccine, 1st dose', 'Japanese encephalitis vaccine, last dose'),
    ('Diphtheria-containing vaccine, 4th dose (1st booster)', 'Diphtheria-containing vaccine, 5th dose (2nd booster)'),
    ('Diphtheria-containing vaccine, 5th dose (2nd booster)', 'Diphtheria-containing vaccine, 6th dose (3rd booster)')
]

# Store results
results = []

# Loop through each pair
for dose1, dose2 in dose_pairs:
    # Get coverage values
    dose1_coverage = data1[data1['ANTIGEN_DESCRIPTION'] == dose1]['COVERAGE'].dropna()
    dose2_coverage = data1[data1['ANTIGEN_DESCRIPTION'] == dose2]['COVERAGE'].dropna()

    # Ensure same length by aligning index (optional; can also just take means)
    avg_dose1 = dose1_coverage.mean()
    avg_dose2 = dose2_coverage.mean()

    # Compute drop-off
    dropoff_pct = ((avg_dose1 - avg_dose2) / avg_dose1) * 100 if avg_dose1 else None

    results.append({
        'DOSE_PAIR': f'{dose1} â†’ {dose2}',
        'AVG_COVERAGE_DOSE1': round(avg_dose1, 2),
        'AVG_COVERAGE_DOSE2': round(avg_dose2, 2),
        'DROP_OFF_%': round(dropoff_pct, 2) if dropoff_pct is not None else None
    })

# Create DataFrame
dropoff_summary = pd.DataFrame(results)

# Display
dropoff_summary

 
```

```python
#3. Are vaccination rates different between genders?

# Merge only on YEAR and NAME
merged_age_coverage = pd.merge(
    data1,
    data5[['NAME', 'YEAR', 'TARGETPOP']],
    on=['NAME', 'YEAR'],
    how='inner'
)

# Group by targetpop and compute average coverage
age_coverage_summary = (
    merged_age_coverage.groupby('TARGETPOP')['COVERAGE']
    .mean()
    .reset_index()
    .sort_values('COVERAGE', ascending=False)
)

# Take only top 10 by average coverage
top10_age_coverage = age_coverage_summary.head(10)

#plot the trend
plt.figure(figsize=(12, 6))
sns.barplot(data=top10_age_coverage, x='TARGETPOP', y='COVERAGE', palette='mako')
plt.xticks(rotation=45)
plt.title('Top 10 Age Groups by Average Vaccination Coverage')
plt.xlabel('Targetpop')
plt.ylabel('Average Coverage (%)')
plt.tight_layout()
plt.grid(True, axis='y')
plt.show()

```

```python
#4. How does education level impact vaccination rates? (considering coverage_category column)

# Group by COVERAGE_CATEGORY and calculate average coverage
category_coverage = (
    data1.groupby('COVERAGE_CATEGORY')['COVERAGE']
    .mean()
    .reset_index()
    .sort_values('COVERAGE', ascending=False)
)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=category_coverage, x='COVERAGE_CATEGORY', y='COVERAGE', palette='Set2')
plt.title('Average Vaccination Coverage by Coverage Category')
plt.xlabel('Coverage Category')
plt.ylabel('Average Coverage (%)')
plt.ylim(0, 100)
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

```python
#5. What is the urban vs. rural vaccination rate difference?

# Calculate average coverage by GROUP 
group_avg = (
    data1.groupby('GROUP')['COVERAGE']
    .mean()
    .reset_index()
    .sort_values(by='COVERAGE', ascending=False)
    .head(30)  # Top 30 for clearer visualization
)

# Plot
plt.figure(figsize=(12, 8))
sns.barplot(data=group_avg, x='COVERAGE', y='GROUP', palette='coolwarm')
plt.title('Average Vaccination Coverage by Group (Country or WHO Region)')
plt.xlabel('Average Coverage (%)')
plt.ylabel('Group')
plt.xlim(0, 100)
plt.grid(True, axis='x')
plt.tight_layout()
plt.show()

```

```python
#6. Has the rate of booster dose uptake increased over time?
# Filter for booster-related antigen descriptions
booster_data = data1[
    data1['ANTIGEN_DESCRIPTION'].str.contains('booster', case=False, na=False)
]

# Calculate average coverage by year and antigen description
booster_trend = booster_data.groupby(['YEAR', 'ANTIGEN_DESCRIPTION'])['COVERAGE'].mean().reset_index()

# Plot the trend for each booster antigen
plt.figure(figsize=(12, 6))
sns.lineplot(data=booster_trend, x='YEAR', y='COVERAGE', hue='ANTIGEN_DESCRIPTION', marker='o')
plt.title('Booster Dose Coverage Over Time by Antigen')
plt.xlabel('Year')
plt.ylabel('Average Coverage (%)')
plt.ylim(0, 100)
plt.legend(title='Booster Antigen', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#7. Is there a seasonal pattern in vaccination uptake?

# Filter for Seasonal Influenza vaccine in DESCRIPTION
influenza_intro = data4[
    data4['DESCRIPTION'].str.contains('Seasonal Influenza', case=False, na=False)
]

# Count number of countries introducing the vaccine per year
intro_by_year = (
    influenza_intro.groupby('YEAR')['NAME']
    .nunique()
    .reset_index()
    .rename(columns={'NAME': 'Countries Introduced'})
)

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=intro_by_year, x='YEAR', y='Countries Introduced', marker='o')
plt.title('Seasonal Influenza Vaccine Introduction Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Countries Introducing')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
#8. How does population density relate to vaccination coverage?

# Merge with data1 to get COVERAGE
merged = pd.merge(
    data2[['NAME', 'YEAR', 'DENOMINATOR']],
    data1[['NAME', 'YEAR', 'COVERAGE']],
    on=['NAME', 'YEAR'],
    how='inner'
)

# Group by denominator type and calculate average coverage
coverage_by_unit = merged.groupby('DENOMINATOR')['COVERAGE'].mean().reset_index()
coverage_by_unit = coverage_by_unit.sort_values('COVERAGE', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
barplot = sns.barplot(data=coverage_by_unit, x='COVERAGE', y='DENOMINATOR', palette='viridis')

# Add percentage labels
for container in barplot.containers:
    barplot.bar_label(container, fmt='%.1f%%', label_type='edge', padding=3)

# Labels and title
plt.xlabel('Average Vaccination Coverage (%)')
plt.ylabel('Denominator Unit')
plt.title('Average Vaccination Coverage by Denominator Unit (Incidence Rate Scale)')
plt.tight_layout()
plt.show()


```

```python
#10. Which regions have high disease incidence despite high vaccination rates?

# Group data by region and calculate the mean incidence rate and coverage
regional_stats = merged_data.groupby('NAME').agg({'INCIDENCE_RATE': 'mean', 'COVERAGE': 'mean'})

# Define thresholds for high incidence and high coverage (adjust as needed)
high_incidence_threshold = np.percentile(regional_stats['INCIDENCE_RATE'], 80)  # Top 20% incidence
high_coverage_threshold = np.percentile(regional_stats['COVERAGE'], 90)  # Top 10% coverage

high_incidence_high_coverage_regions = regional_stats[
    (regional_stats['INCIDENCE_RATE'] > high_incidence_threshold) &
    (regional_stats['COVERAGE'] > high_coverage_threshold)
]

print("Regions with high disease incidence despite high vaccination rates:")
high_incidence_high_coverage_regions
```

```python
#1. Is there a correlation between vaccine introduction and a decrease in disease cases?

# Merge the vaccine introduction data with disease incidence data based on year and region
merged_vaccine_disease_data = pd.merge(data3, data4, on=['YEAR', 'NAME'], how='inner')

# Convert the 'VACCINE_INTRODUCTION' column to binary (1 for 'yes', 0 for 'no')
merged_vaccine_disease_data['INTRO'] = merged_vaccine_disease_data['INTRO'].map({'Yes': 1, 'No': 0})

# Calculate the correlation between vaccine introduction and the reduction in disease cases.
correlation_vaccine_disease = merged_vaccine_disease_data['INTRO'].corr(merged_vaccine_disease_data['CASES'])

print(f"Correlation between vaccine introduction and disease cases reduction: {correlation_vaccine_disease}")

# Visualize the relationship between vaccine introduction and disease cases reduction.
plt.figure(figsize=(8, 6))
sns.scatterplot(x='INTRO', y='CASES', data=merged_vaccine_disease_data)
plt.xlabel('Vaccine Introduction (1 = Yes, 0 = No)')
plt.ylabel('Reported Disease Cases')
plt.title('Vaccine Introduction vs Disease Cases Reduction')
plt.show()

# For specific diseases (optional), check the correlation for each disease type (replace 'DISEASE' with your disease column).
for disease in merged_vaccine_disease_data['DISEASE'].unique():
    disease_data = merged_vaccine_disease_data[merged_vaccine_disease_data['DISEASE'] == disease]
    disease_correlation = disease_data['INTRO'].corr(disease_data['CASES'])
    print(f"Correlation for {disease}: {disease_correlation}")

```

```python
#2. What is the trend in disease cases before and after vaccination campaigns?

# Filter for years
filtered = data3[data3['YEAR'].isin([1980, 1990])]

# Group by disease and year, sum cases
grouped = filtered.groupby(['DISEASE', 'YEAR'])['CASES'].sum().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=grouped, x='DISEASE', y='CASES', hue='YEAR')
plt.title('Disease Cases in 1980 vs 1990')
plt.xlabel('Disease')
plt.ylabel('Reported Cases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(True, axis='y')
plt.show()
```

```python
#3. Which diseases have shown the most significant reduction in cases due to vaccination?


# Filter data for years
filtered = data3[data3['YEAR'].isin([1980, 2000])]

# Group by disease and year, sum reported cases
grouped = filtered.groupby(['DISEASE', 'YEAR'])['CASES'].sum().unstack()

# Calculate % reduction
grouped['PERCENT_REDUCTION'] = ((grouped[1980] - grouped[2000]) / grouped[1980]) * 100

# Drop diseases with missing data
grouped = grouped.dropna()

# Sort by percent reduction descending
reduced_diseases = grouped.sort_values(by='PERCENT_REDUCTION', ascending=False)

# Show top 5
print("Top 5 diseases with the highest reduction in cases (1980 to 2000):")
print(reduced_diseases[['PERCENT_REDUCTION']].head(5).round(2))
```

```python
#7. How does vaccine coverage correlate with disease reduction for specific antigens?

# Filter for selected antigens
selected_antigens = ['BCG', 'HPV_FEM', 'PAB']
filtered_cov = data1[data1['ANTIGEN'].isin(selected_antigens)]

# Prepare disease data
filtered_dis = data2[['NAME', 'YEAR', 'DISEASE', 'INCIDENCE_RATE']]

# Merge datasets
merged = pd.merge(filtered_cov, filtered_dis, on=['NAME', 'YEAR'], how='inner')

# Plot and correlation for each antigen
for antigen in selected_antigens:
    sub = merged[merged['ANTIGEN'] == antigen]
    if not sub.empty:
        plt.figure(figsize=(8, 5))
        sns.scatterplot(data=sub, x='COVERAGE', y='INCIDENCE_RATE')
        sns.regplot(data=sub, x='COVERAGE', y='INCIDENCE_RATE', scatter=False, color='red')
        plt.title(f'{antigen}: Coverage vs Disease Incidence')
        plt.xlabel('Vaccination Coverage (%)')
        plt.ylabel('Disease Incidence Rate')
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        # Correlation
        corr = sub['COVERAGE'].corr(sub['INCIDENCE_RATE'])
        print(f'Correlation for {antigen}: {corr:.2f}')
    else:
        print(f"No data available for {antigen}")
```

```python
#8. Are there specific regions or countries with low coverage despite high availability of vaccines?

# Filter for countries with high vaccine availability (introduced)
available = data4[data4['INTRO'].str.upper() == 'YES'][['NAME', 'DESCRIPTION', 'YEAR']].drop_duplicates()

# Merge with coverage data
merged_avail_cov = pd.merge(
    available,
    data1[['NAME', 'YEAR', 'COVERAGE']],
    on=['NAME', 'YEAR'],
    how='left'
)

# Find countries/regions with low coverage (<70%) despite high availability
low_coverage = merged_avail_cov[merged_avail_cov['COVERAGE'] < 70]

# Plot (Top 20 by lowest coverage)
top_low = (
    low_coverage.groupby('NAME')['COVERAGE']
    .mean()
    .sort_values()
    .head(20)
    .reset_index()
)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_low, x='COVERAGE', y='NAME', palette='Reds_r')
plt.title('Countries/Regions with Low Coverage Despite High Vaccine Availability')
plt.xlabel('Average Coverage (%)')
plt.ylabel('Country/Region')
plt.xlim(0, 100)
plt.tight_layout()
plt.show()
```

```python
#9. What are the gaps in coverage for vaccines targeting high-priority diseases (e.g., TB, Hepatitis B)?

# Define high-priority antigens 
high_priority_antigens = ['BCG', 'HEPB3', 'HEPB_BD']

# Filter data for these antigens
priority_coverage = data1[data1['ANTIGEN'].isin(high_priority_antigens)]

# Calculate average coverage by country and antigen
coverage_gaps = (
    priority_coverage.groupby(['NAME', 'ANTIGEN'])['COVERAGE']
    .mean()
    .reset_index()
)

# Take top 15 countries with lowest average coverage (across all antigens)
low_coverage_countries = (
    coverage_gaps.groupby('NAME')['COVERAGE']
    .mean()
    .sort_values()
    .head(15)
    .index
)

filtered_gaps = coverage_gaps[coverage_gaps['NAME'].isin(low_coverage_countries)]

plt.figure(figsize=(14, 7))
sns.barplot(
    data=filtered_gaps,
    x='COVERAGE',
    y='NAME',
    hue='ANTIGEN',
    palette='Set2'
)
plt.title('Coverage Gaps for High-Priority Vaccines (Lowest 15 Countries)')
plt.xlabel('Average Coverage (%)')
plt.ylabel('Country/Region')
plt.xlim(0, 100)
plt.legend(title='Vaccine (Antigen)')
plt.tight_layout()
plt.show()
```

```python
#10. Are certain diseases more prevalent in specific geographic areas? 

# Group by region and disease, calculate average incidence rate
region_disease = (
    data2.groupby(['NAME', 'DISEASE'])['INCIDENCE_RATE']
    .mean()
    .reset_index()
)

# Pivot for heatmap: rows=regions, columns=diseases, values=incidence rate
heatmap_data = region_disease.pivot(index='NAME', columns='DISEASE', values='INCIDENCE_RATE')

# top N regions with highest total incidence
top_regions = heatmap_data.sum(axis=1).sort_values(ascending=False).head(20).index
heatmap_data = heatmap_data.loc[top_regions]

plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, annot=False, cmap='Reds', linewidths=0.5, cbar_kws={'label': 'Avg Incidence Rate'})
plt.title('Disease Prevalence by Region (Top 20 Regions)')
plt.xlabel('Disease')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

```python
#2. A public health organization wants to evaluate the effectiveness of a measles vaccination campaign launched five years ago.

# Compare measles vaccination coverage and reported cases between 1980 and 2023

# Measles vaccination coverage
measles_cov = data1[data1['ANTIGEN_DESCRIPTION'].str.contains('Measles', case=False, na=False)]
cov_years = measles_cov[measles_cov['YEAR'].isin([1980, 2023])]
avg_cov = cov_years.groupby('YEAR')['COVERAGE'].mean().reset_index()

# Measles reported cases
measles_cases = data3[data3['DISEASE'].str.lower() == 'measles']
cases_years = measles_cases[measles_cases['YEAR'].isin([1980, 2023])]
total_cases = cases_years.groupby('YEAR')['CASES'].sum().reset_index()

# Plot coverage
plt.figure(figsize=(8, 4))
sns.barplot(data=avg_cov, x='YEAR', y='COVERAGE', palette='Blues')
plt.title('Average Measles Vaccination Coverage (1980 vs 2023)')
plt.ylabel('Coverage (%)')
plt.xlabel('Year')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()

# Plot cases
plt.figure(figsize=(8, 4))
sns.barplot(data=total_cases, x='YEAR', y='CASES', palette='Reds')
plt.title('Total Reported Measles Cases (1980 vs 2023)')
plt.ylabel('Number of Cases')
plt.xlabel('Year')
plt.tight_layout()
plt.show()
```

```python
#3. A vaccine manufacturer wants to estimate vaccine demand for a specific disease in the upcoming year.

# Filter for measles coverage and cases
measles_cov = data1[data1['ANTIGEN_DESCRIPTION'].str.contains('Measles', case=False, na=False)]
measles_cases = data3[data3['DISEASE'].str.lower() == 'measles']

# Aggregate by year
coverage_trend = measles_cov.groupby('YEAR')['COVERAGE'].mean().reset_index()
cases_trend = measles_cases.groupby('YEAR')['CASES'].sum().reset_index()

# Plot: Average coverage per year
plt.figure(figsize=(16, 4))
sns.lineplot(data=coverage_trend, x='YEAR', y='COVERAGE', marker='o', color='green')
plt.title('Measles: Average Vaccination Coverage per Year')
plt.ylabel('Average Coverage (%)')
plt.xlabel('Year')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()

# Plot: Total reported cases per year
plt.figure(figsize=(16, 4))
sns.barplot(data=cases_trend, x='YEAR', y='CASES', color='salmon')
plt.title('Measles: Total Reported Cases per Year')
plt.ylabel('Reported Cases')
plt.xlabel('Year')
plt.tight_layout()
plt.show()
```

```python
#5. Researchers want to explore the incidence rates of polio in populations with no vaccination coverage.

# Merge coverage and incidence data on region and year
merged = pd.merge(
    data1[['NAME', 'YEAR', 'ANTIGEN_DESCRIPTION', 'COVERAGE']],
    data2[['NAME', 'YEAR', 'DISEASE', 'INCIDENCE_RATE']],
    left_on=['NAME', 'YEAR'],
    right_on=['NAME', 'YEAR'],
    how='inner'
)

# Filter for polio disease and zero coverage
polio_no_coverage = merged[
    (merged['DISEASE'].str.contains('polio', case=False, na=False)) &
    (merged['ANTIGEN_DESCRIPTION'].str.contains('polio', case=False, na=False)) &
    (merged['COVERAGE'] == 0)
]

# Show relevant columns
result = polio_no_coverage[['NAME', 'YEAR', 'INCIDENCE_RATE']].sort_values(['INCIDENCE_RATE'], ascending=False)
print("Incidence rates of polio in populations with no vaccination coverage:")
print(result.head(20))

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=result.head(10), x='INCIDENCE_RATE', y='NAME', hue='YEAR', dodge=False, palette='Reds')
plt.title('Top Regions/Years: Polio Incidence Rate with No Vaccination Coverage')
plt.xlabel('Polio Incidence Rate')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

```python
#6. WHO wants to track global progress toward achieving a target of 95% vaccination coverage for measles by 2030.

# Filter for measles vaccination coverage
measles_cov = data1[data1['ANTIGEN_DESCRIPTION'].str.contains('measles', case=False, na=False)]

# Group by year and calculate global average coverage
global_measles_coverage = measles_cov.groupby('YEAR')['COVERAGE'].mean().reset_index()

# Plot the trend and the 95% target line
plt.figure(figsize=(12, 6))
sns.lineplot(data=global_measles_coverage, x='YEAR', y='COVERAGE', marker='o', label='Global Average Coverage')
plt.axhline(95, color='red', linestyle='--', label='WHO Target (95%)')
plt.title('Global Measles Vaccination Coverage Over Time')
plt.xlabel('Year')
plt.ylabel('Average Coverage (%)')
plt.ylim(0, 100)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#7. A health agency wants to allocate vaccines to high-risk populations such as children under five and the elderly.

# Only consider rows where TARGETPOP is 'riskgroups'
risk_groups = data5[data5['TARGETPOP'].str.lower() == 'riskgroups']

# Merge with coverage data to get vaccination coverage for these groups
risk_coverage = pd.merge(
    risk_groups[['NAME', 'YEAR', 'TARGETPOP']],
    data1[['NAME', 'YEAR', 'COVERAGE']],
    on=['NAME', 'YEAR'],
    how='left'
)

# Group by region and year to see coverage trends for riskgroups
riskgroup_coverage_trend = (
    risk_coverage.groupby(['NAME', 'YEAR'])['COVERAGE']
    .mean()
    .reset_index()
    .sort_values(['YEAR', 'COVERAGE'], ascending=[True, False])
)

print("Vaccination coverage for high-risk groups by region and year:")
print(riskgroup_coverage_trend.head(20))

# Plot: Average coverage for riskgroups by year (global trend)
plt.figure(figsize=(10, 6))
global_riskgroup_trend = risk_coverage.groupby('YEAR')['COVERAGE'].mean().reset_index()
sns.lineplot(data=global_riskgroup_trend, x='YEAR', y='COVERAGE', marker='o')
plt.title('Global Vaccination Coverage for High-Risk Groups Over Time')
plt.xlabel('Year')
plt.ylabel('Average Coverage (%)')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()

# Plot: Top 10 regions by average coverage for riskgroups
top_regions = (
    risk_coverage.groupby('NAME')['COVERAGE']
    .mean()
    .reset_index()
    .sort_values('COVERAGE', ascending=False)
    .head(10)
)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_regions, x='COVERAGE', y='NAME', palette='crest')
plt.title('Top 10 Regions by Vaccination Coverage for High-Risk Groups')
plt.xlabel('Average Coverage (%)')
plt.ylabel('Region')
plt.xlim(0, 100)
plt.tight_layout()
plt.show()
```

```python
#8. A non-profit wants to detect disparities in vaccination coverage across different socioeconomic groups within a country.

# Group by country (NAME) and COVERAGE_CATEGORY, then calculate average coverage
coverage_by_category_country = (
    data1.groupby(['NAME', 'COVERAGE_CATEGORY'])['COVERAGE']
    .mean()
    .reset_index()
    .sort_values(['NAME', 'COVERAGE_CATEGORY'])
)


# Plot: Heatmap of coverage by country and coverage category (top 10 countries for clarity)
top_countries = coverage_by_category_country['NAME'].value_counts().head(10).index
heatmap_data = coverage_by_category_country[coverage_by_category_country['NAME'].isin(top_countries)]
heatmap_pivot = heatmap_data.pivot(index='NAME', columns='COVERAGE_CATEGORY', values='COVERAGE')

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_pivot, annot=True, fmt=".1f", cmap='YlGnBu', cbar_kws={'label': 'Avg Coverage (%)'})
plt.title('Vaccination Coverage by Socioeconomic Group (Coverage Category) and Country')
plt.xlabel('Coverage Category')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
